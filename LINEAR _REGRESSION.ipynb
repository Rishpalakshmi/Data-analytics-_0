```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Load dataset
data = pd.read_csv("salary.csv")


# Step 3: Separate input and output
X = data[['Years']]      # Independent variable
y = data['Salary']       # Dependent variable

# Step 4: Create model
model = LinearRegression()

# Step 5: Train model
model.fit(X, y)

# Step 6: Predict salary for 9 years experience
prediction = model.predict([[9]])
print("Predicted Salary for 9 years experience:", prediction[0])

# Step 7: Plot graph
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression using CSV Dataset")
plt.show()
```
